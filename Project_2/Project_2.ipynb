```python
import pandas as pd
```

#We create a dataframe

```python

#Creating the dataframe
names = ["Chapman, Matt", "Olson, Matt", "Davis, Khris"]
positions = ["3B", "1B", "OF"]
hrs = [25,30,45]
df_plyrs = pd.DataFrame({"names":names, "pos":positions, "hr":hrs})
df_plyrs
```

#We use apply function to add a column for each players last name.


```python
#Write your code here

def Get_lastN(column):
    name = column.split(" ")
    
    for i in range(len(name)):
        if "," in name[i]:
            return name[i]
    return column

df_plyrs["LastN"] = df_plyrs["names"].apply(Get_lastN)
df_plyrs
```

#We add a column showing if the player is an all-star category. We say a player is an all-star if they play in a third base position or outfield. Our threshold_hr = 20 homeruns. If the player is an all-star, then the row in the new column should become "all-star" otherwise it must be "not selected".




```python
def allstart(df):
    position = df['pos']
    threshold_hr = df["hr"]
    if position == "3B" and threshold_hr>=20:
        return 'Yes'
    else:
        return 'No'
    
df_plyrs["All Star"] = df_plyrs.apply(allstart, axis=1)
df_plyrs
```

```python
df = pd.read_csv("Movie_Reviews-3 2.tsv", sep='\t')
df.head()
```

```python
def Sentiment(df):
    
    sentiment = df['Sentiment']
    if sentiment == 0:
        return 'Negative'
    elif sentiment == 1:
        return 'Somewhat Negative' 
    elif sentiment == 2:
        return 'Neutral' 
    elif sentiment == 3:
        return 'Somewhat Positive' 
    else:
        return 'Positive' 
    
df['Sentiment_Meanings'] = df.apply(Sentiment,axis=1)
df
```

#we get rid of all phrases that have fewer than 10 words

```python
#Add phrase_length column

def count_phrase(df):
    phrase = df['Phrase']
    
    words = phrase.split(" ")
    
    return len(words)

df['Phrase Count'] = df.apply(count_phrase, axis=1)
```

```python
df.tail()
```

```python
#Get ride of phrases that have fewer than 10 words
def get_rid_10(df):
    df.drop(df[df['Phrase Count'] < 10].index, inplace = True)

get_rid_10(df)
df
```

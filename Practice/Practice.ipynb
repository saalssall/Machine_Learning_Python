```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def graient_descent(x, y):
    
    #Step 1: we initially take random values for m and b==0
    
    m_curr =0
    b_curr =0
    
    #Step 2: n is the total number of x
    
    n=len(x)
    
    #Step 3: we get the learning rate value
    
    rate =0.1
    
    #Step 4: we set iterations
    
    iterNum=1000
    
    #Step 5: plot the graph
    
    plt.scatter(x, y, color="k")
    
    m_curr_history=np.zeros(iterNum)
    b_curr_history=np.zeros(iterNum)
    
    for i in range(iterNum):
        
        yp = m_curr*x + b_curr   # y is the predicted value
        
        plt.plot(x, yp, color="yellow")
    
        #sum square of each delta value(y-y^)
    
        md = -(2/n)*sum(x*(y-yp))
        bd= -(2/n)*sum(y-yp)
        
        #We update our m_curr and b_curr
        
        m_curr= m_curr- rate*md
        b_curr= b_curr- rate*bd
        
        mse = m_curr*x + b_curr
        
        
        m_curr_history[i]=m_curr
        
        b_curr_history[i]=b_curr
        
    return m_curr_history, b_curr_history, mse
       
    
    
    
    
x =np.array([1,2,3,4,5])
y =np.array([5, 7, 9, 11, 13])

m_curr_history, b_curr_history, mse = graient_descent(x, y)

#print("All the md  values are: ", md_history) #this shows the values of m or slope in the 1000 iteration

plt.plot(x, yp, color="red") #shows the best fit line after updating m_curr and b_curr
```
